# Paper Folding Problem by Wanderley

## Problem 1: Implementation using exponentiation arithmetic operators

```python
#Folded once
THICKNESS = 0.00008
folded_thickness = THICKNESS * 2

print("Thickness: {} meters".format(folded_thickness))
```

```python
#Folded 43 times
folded_thickness = THICKNESS * (2** 43)

print("Thickness: {} meters".format(folded_thickness))
```

## Problem 2: Unit conversion

```python
print("Thickness: {: .2f} kilometers".format(folded_thickness/1000))
```

## Problem 3: Created using a for statement

```python
folded_thickness = THICKNESS

for i in range (43):
    folded_thickness *= 2
    
print("Thickness: {} meters".format(folded_thickness))
```

## Problem 4: Comparison of calculation times

```python
#Using time.time() function
import time

start = time.time()
folded_thickness = THICKNESS * (2** 1000)
elapsed_time = time.time() - start
print("first time : {:.9f}[s]".format(elapsed_time))

start = time.time()
for i in range (1000):
    folded_thickness *= 2
elapsed_time = time.time() - start
print("second time : {:.9f}[s]".format(elapsed_time))
```

```python
#Using time.perf_counter() function for higher precision

start = time.perf_counter()
folded_thickness = THICKNESS * (2 ** 43)
elapsed_time = time.perf_counter() - start
print("First method time: {:.9f}[s]".format(elapsed_time))

# Using a loop
folded_thickness = THICKNESS  # Reset the thickness

start = time.perf_counter()
for i in range(43):
    folded_thickness *= 2
elapsed_time = time.perf_counter() - start
print("Second method time: {:.9f}[s]".format(elapsed_time))
```

## Problem 5: Saving to a list

```python
folded_list = [THICKNESS]
for i in range(44):
    if i == 0:
        continue
        
    folded_list.append(folded_list[i-1] * 2)
    
print(folded_list)
print(f"The length of the list is {len(folded_list)} ")
```

## Problem 6: Displaying a line graph

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(folded_list)
plt.show()
```
